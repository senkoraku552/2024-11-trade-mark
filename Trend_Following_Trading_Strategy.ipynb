```python
import yfinance as yf
import pandas as pd
import numpy as np

# Load historical data for the asset you want to trade

data = yf.download("0003.HK", start="2023-06-01", end="2023-10-27")

data["Close"].tail()
```

```python
# Calculate the moving average of the asset's price over a certain period of time
data["ma"] = data["Close"].rolling(window=20).mean()
data.tail()

```

```python
# Create a new column that shows whether the asset's price is trending upward or downward
data["trend"] = np.where(data["Close"] > data["ma"], 1, -1)
data["trend"].tail()
```

```python
# Use the trend column to make buy or sell decisions
#for i in range(len(data)):

print("Trend Following Trading Strategy")
Buy_time=1.0
data2=data.loc[data.trend==Buy_time]
print("Long")
print(data2["Close"])

print("")
print("")
Sell_time= -1.0
data2=data.loc[data.trend==Sell_time]
print("Short")
print(data2["Close"])
   
```
